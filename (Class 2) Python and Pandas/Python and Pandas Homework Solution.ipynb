```python
import requests
import pandas as pd

url = 'http://data.cityofnewyork.us/api/views/jb7j-dtam/rows.json'
results = requests.get(url).json() # reading in the json just as we did with our citibike info last week
```

```python
results = requests.get(url).json() # reading in the json just as we did with our citibike info last week
# there are two main fields returned in the json, the meta that just describes the actual metadata, and the 
# data itself

df = pd.DataFrame(results["data"])
df.head()
```

```python
columns = results["meta"]["view"]["columns"] # this gives us the descriptions and names for the columns

headers = [i["fieldName"] for i in columns] # we create a list of the column names 

df = pd.DataFrame(results["data"], columns=headers) # now we pass in a list of those column names to our df
```

```python
# hint

df.drop(labels = [':sid', ':position', ':meta', ':created_meta', ':updated_meta'], axis='columns', inplace=True)
```

```python
# drop those first three rows as they appear to be metadata that for some reason were included in our df

df = df.drop([0,1,2],axis=0)
```

```python
# leading cause of death in 2012 

df['year'] = pd.to_numeric(df["year"])

df2012 = df[df['year'] == 2012]

df2012['leading_cause'].describe()
```

```python
# what race/ethnicity had the most recorded deaths in 2013? 

df["deaths"] = pd.to_numeric(df["deaths"], errors='coerce')

df2013 = df[df['year'] == 2013]

df2013.groupby(['race_ethnicity']).sum()
```

```python
# year with lowest age adjusted death rate 

df["age_adjusted_death_rate"] = pd.to_numeric(df["age_adjusted_death_rate"], errors='coerce')

df.groupby(['age_adjusted_death_rate'],sort=True).sum()
```
