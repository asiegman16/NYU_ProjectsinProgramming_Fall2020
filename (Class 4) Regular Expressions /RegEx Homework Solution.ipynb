```python
import requests
url = 'http://www.stern.nyu.edu/faculty/search_name_form'

html = requests.get(url).text
```

```python
import re

emails = set() # set doesn't allow for duplicates, FYI

pattern = r'\w+@\w+\.nyu\.edu'
regex = re.compile(pattern)
matches = regex.finditer(html)
for m in matches:
    emails.add(m.group())

sorted(emails)
```
